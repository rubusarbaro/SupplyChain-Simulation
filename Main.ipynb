```python
class Material:
    def __init__(self,pn: str, description: str):
        self.pn = pn
        self.description = description

class Raw_Material(Material):
    def __init__(self, pn: str, description: str):
        super().__init__(pn, description)

class Assembly(Material):
    def __init__(self, pn: str, description: str, bom: list, rate_ps: int):
        super().__init__(pn, description)
        self.rate_ps = rate_ps

class Final_Product(Assembly):
    def __init__(self, pn: str, description: str, bom: list, rate_ps: int):
        super().__init__(pn, description, bom, rate_ps)

class Production_Line:
    def __init__(self, line_name: str, product: str, rate_ps: int):
        self.name = line_name
        self.product = product
        self.rate_ps = rate_ps
        self.finished_goods = 0

    def start(self):
        self.finished_goods += self.rate_ps
        print(f"{self.name} ha producido {self.finished_goods} {self.product}")
```

```python
Torno = Production_Line("Línea 1", "tornillos", 100)
```

```python
clock = 0
while True:
    output.clear()
    Torno.start()
    sleep(1)
```
