```python
from google.colab import output
from time import sleep
```

```python
class Production_Line:
    def __init__(self, line_name: str, product: str, rate_ps: int):
        self.name = line_name
        self.product = product
        self.rate_ps = rate_ps
        self.finished_goods = 0

    def start(self):
        self.finished_goods += self.rate_ps
        print(f"{self.name} ha producido {self.finished_goods} {self.product}")
```

```python
Torno = Production_Line("Línea 1", "tornillos", 100)
```

```python
clock = 0
while True:
    output.clear()
    Torno.start()
    sleep(1)
```
